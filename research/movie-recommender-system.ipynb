```python
movies = pd.read_csv("/kaggle/input/moviesdata/tmdb_5000_movies.csv")
credits = pd.read_csv("/kaggle/input/moviesdata/tmdb_5000_credits.csv")
```

```python
print(f"shape of the movies: {movies.shape}")
print(f"shape of the credits: {credits.shape}")
```

```python
## merging both of the datasets because we don't want two datasets
movies = movies.merge(credits, on="title")
movies.drop(columns=["movie_id"], axis=1, inplace=True)
movies.head()
```

```python
print(f"shape of the movies: {movies.shape}")
```

```python
## dropping the columns that not helps us
cols_to_drop = ["budget", "homepage", "original_language", "original_title", "popularity", "production_companies", "production_countries",
                "release_date", "revenue", "runtime", "spoken_languages", "status", "tagline", "vote_average", "vote_count"]

movies.drop(columns=cols_to_drop, axis=1, inplace=True)
movies.head()
```

```python
## now we're gonna create a new dataframe in which we have mainly three columns id, title, tags (combination of overview, genres, keywords, cast, crew)
## data preprocessing of all the tags columns
## checking for missing values

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
## checking for duplicate values 
movies.duplicated().sum()
```

```python
## getting the genres, keywords, overview, cast and crew columns in the right format
movies["genres"][0]
```

```python
import ast

def helper_fun(obj):
    lst = []
    for x in ast.literal_eval(obj):
        lst.append(x["name"])
    
    return lst
```

```python
movies["genres"] = movies["genres"].apply(helper_fun)
movies["genres"][0]
```

```python
movies["keywords"] = movies["keywords"].apply(helper_fun)
movies.head()
```

```python
def helper_fun2(obj):
    lst = []
    counter = 0
    for x in ast.literal_eval(obj):
        if counter != 4:
            lst.append(x["name"])
            counter+=1
        else:
            break
            
    return lst
```

```python
movies["cast"] = movies["cast"].apply(helper_fun2)
movies["cast"][0]
```

```python
def helper_fun3(obj):
    lst = []
    for x in ast.literal_eval(obj):
        if x["job"] == "Director":
            lst.append(x["name"])
            break
            
    return lst
```

```python
movies["crew"] = movies["crew"].apply(helper_fun3)
movies["crew"][0]
```

```python
## converting the overview column into list
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
## now we have the data in the desired format
movies.head()
```

```python
## we need to remove the spaces between the each word in the list for the columns genres, keywords, cast and crew to avoid the confussion for our recommendation system
movies["genres"] = movies["genres"].apply(lambda x : [item.replace(" ", "") for item in x])
movies["keywords"] = movies["keywords"].apply(lambda x : [item.replace(" ", "") for item in x])
movies["cast"] = movies["cast"].apply(lambda x : [item.replace(" ", "") for item in x])
movies["crew"] = movies["crew"].apply(lambda x : [item.replace(" ", "") for item in x])

movies.head()
```

```python
## creating a tags columns by adding the genres, overview, keywords, cast and crew columns
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
movies.drop(columns=["overview", "genres", "keywords", "cast", "crew"], axis=1, inplace=True)

movies.head()
```

```python
## converting the tags into string 
movies["tags"] = movies["tags"].apply(lambda x : " ".join(x))

## converting into the lower case
movies["tags"] = movies["tags"].apply(lambda x : x.lower())
```

```python
movies.head()
```

```python
## the problem over here, we want the words 'action' and 'actions' (all of the similar words) to be interpreted as same
## for this we're gonna use stemming
from nltk.stem import PorterStemmer

stem_obj = PorterStemmer()
```

```python
def stemmer(text):
    lst = []
    for x in text.split():
        lst.append(stem_obj.stem(x))
        
    return " ".join(lst)
```

```python
movies["tags"] = movies["tags"].apply(stemmer)
movies.head()
```

```python
## now our need is to do vectorization of the tags so that we can extract most closest or similar kind of movies
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=6500, stop_words='english')

vectors = vectorizer.fit_transform(movies["tags"]).toarray()
```

```python
print(f"shape of the vectors: {vectors.shape}\n")
print(vectors)
```

```python
## we're gonna use cosine_similarity instead of euclidean distance because it fails when the dimensions are high
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
print(f"shape of the distances: {similarity.shape}")
```

```python
similarity[0]
```

```python
## now here comes the heart of the recommendation system: main function
def recommend(movie):
    movie_inx = movies[movies["title"] == movie].index[0]
    distances = similarity[movie_inx]
    similar_movies = sorted(enumerate(distances), reverse=True, key=lambda x : x[1])[1:5]
    
    for x in similar_movies:
        print(movies.iloc[x[0]].title,"\n")
```

```python
recommend("Batman Begins")
```
